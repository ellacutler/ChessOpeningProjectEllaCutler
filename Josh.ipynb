```python
def remove_chunk(raw_game, s0, e0, s1, e1, s2, e2, st0, st1, st2):
    result = ""
    for i in range(len(raw_game)):
        char = raw_game[i]
        if char == s0:
            st0 = False
        if char == s1:
            st1 = False
        if char == s2:
            st2 = False
        if st0 == True and st1 == True and st2 == True:
            result += char
        if char == e0:
            st0 = True
        if char == e1:
            st1 = True
        if char == e2:
            st2 = True
    return result.strip()

def remove_nums(input):
    result = ""
    for i in range(len(input) - 1):
        char = input[i+1]
        bchar = input[i]
        if char.isdigit() and bchar.isalpha() == False:
            continue
        else:
            result += char
    return result.strip()

def pgn_add_slash(input):
    result = '/'.join(input.split()[:-1])
    return "SoG/" + result
        


```

```python
def calc_master_moves(inp):

    slcount = inp.count('/')
    sl1 = slcount + 2
    movedict = {}
    
    list = MasterPrep.keys(prefix=inp)
                    
    for i in range(len(list)):
        current = list[i]
        nmov = "/".join(current.split("/", sl1)[:sl1])
        res = nmov.rsplit('/', 1)[1]
        if res in movedict:
            movedict[res] += 1
        else:
            movedict[res] = 1
    
    return movedict
        
```
